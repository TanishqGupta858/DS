### Now I will collect data by Webscraping 

##### For web scraping we need few libraries like pandas, beautifulsoup and requests

```python
from bs4 import BeautifulSoup
```

```python
url = "https://en.wikipedia.org/wiki/Electoral_history_of_the_Bharatiya_Janata_Party"
```

```python
page = requests.get(url)
```

```python
soup = BeautifulSoup(page.text,'html')
```

```python
soup
```

```python
table = soup.find_all('table')[1]
print(table)
```

```python
title_table = table.find_all('th')
title_table
```

```python
table_title = [title.text.strip() for title in title_table]
table_title
```

```python
df = pd.DataFrame(columns = table_title)
df
```

```python
columns_data = table.find_all('tr')
```

```python
for row in columns_data[1: ]:
    row_data = row.find_all('td')
    ind_row_data = [data.text.strip() for data in row_data]
    length = len(df)
    df.loc[length] = ind_row_data
```
