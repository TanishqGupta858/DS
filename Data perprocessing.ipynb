```python
df = df.reset_index()
df
```

```python
df = df.drop(['index'], axis=1)
df
```

```python
sns.boxplot(x = df["BHK_NO."])
```

```python
Q1 = df['BHK_NO.'].quantile(0.25)
Q3 = df['BHK_NO.'].quantile(0.75)
IQR = Q3 - Q1
lower = Q1 - 1.5*IQR
upper = Q3 + 1.5*IQR

upper_array = np.where(df['BHK_NO.'] >= upper)[0]
lower_array = np.where(df['BHK_NO.'] <= lower)[0]

df.drop(index=upper_array, inplace=True)
df.drop(index=lower_array, inplace=True)
```

```python
df.shape
```

```python
df = df.reset_index()
df = df.drop(['index'], axis=1)
df
```

```python
sns.boxplot(data = df["SQUARE_FT"])
```

```python
Q1 = df['SQUARE_FT'].quantile(0.25)
Q3 = df['SQUARE_FT'].quantile(0.75)
IQR = Q3 - Q1
lower = Q1 - 1.5*IQR
upper = Q3 + 1.5*IQR

upper_arrays = np.where(df['SQUARE_FT'] >= upper)[0]
lower_arrays = np.where(df['SQUARE_FT'] <= lower)[0]

df.drop(index=upper_arrays, inplace=True)
df.drop(index=lower_arrays, inplace=True)
```

```python
df.shape
```

```python
df = df.reset_index()
df = df.drop(['index'], axis=1)
df
```

```python
sns.boxplot(data = df['TARGET(PRICE_IN_LACS)'])
```

```python
Q1 = df['TARGET(PRICE_IN_LACS)'].quantile(0.25)
Q3 = df['TARGET(PRICE_IN_LACS)'].quantile(0.75)
IQR = Q3 - Q1
lower = Q1 - 1.5*IQR
upper = Q3 + 1.5*IQR

upper_arrays = np.where(df['TARGET(PRICE_IN_LACS)'] >= upper)[0]
lower_arrays = np.where(df['TARGET(PRICE_IN_LACS)'] <= lower)[0]

df.drop(index=upper_arrays, inplace=True)
df.drop(index=lower_arrays, inplace=True)
```

```python
df.shape
```

```python
df = df.reset_index()
df = df.drop(['index'], axis=1)
df
```
